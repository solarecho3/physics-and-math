```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(x):
    return 2*x**3 - 5*x + 7

def df(x):
    return 6*x**2 - 5

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)
y = f(x)
dy = df(x)

plt.figure(figsize=(12,5))

# Plot the function f(x) and its derivative df(x)
plt.subplot(1,2,1)   # 1 row, 2 columns, first plot
plt.plot(x, y)
plt.title('f(x) = 2*x^3 - 5*x + 7')
plt.grid()

plt.subplot(1,2,2)   # 1 row, 2 columns, second plot
plt.plot(x, dy)
plt.title('f\'(x) = df/dx = 6*x^2 - 5')
plt.grid()

# Show the plots
plt.show()
```

# The Concept of Derivative

A **derivative** is one of the main ideas in calculus that measures how a function changes as its input changes. It can be viewed as an instantaneous rate of change or slope at any 
point on the graph of the function. This concept is fundamental to understanding many phenomena, from physics and engineering to economics.

Considering our functions $$f(x) = 2*x^3 - 5*x + 7$$ and its derivative $$f'(x) = df/dx = 6*x^2 - 5$$
- The function **f(x)** represents a curve on the graph. At any point along this curve, you can imagine drawing a tangent (a line that just 'touches' the curve at that point). The 
slope of that tangent line is what we call the derivative at that particular point.

- The function **f'(x)** represents another curve, which shows how fast or slow the original curve (f(x)) is changing at any given x value. If you follow a point along this f'(x) 
curve from left to right, it's like watching how the slope of our tangent line on the f(x) plot changes - that's exactly what the derivative tells us!

# Limits

## Simple

A limit is defined as $\lim_{x \to a} f(x)$

Specifically, $\lim_{x \to 2} x^2 = 4$

```python
def square(x):
    return x**2
linspace = np.linspace(0,2,100)
y = [square(x) for x in linspace]
plt.plot(linspace,y)
```
