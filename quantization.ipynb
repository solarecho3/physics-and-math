```python
import numpy as np
import matplotlib.pyplot as plt
# from scipy.interpolate import interp

# Parameters
fs = 44100  # Sampling frequency (Hz)
T = 1 # Duration of signal (seconds)
frequency = 5  # Frequency of sine wave (Hz)
amplitude = 1  # Amplitude of sine wave
num_samples = fs * T  # Total number of samples
time = np.linspace(0, T, num_samples, endpoint=False)  # Time axis
sine_wave = amplitude * np.sin(2 * np.pi * frequency * time)
```

```python
sine_wave
```

```python
plt.plot(sine_wave)
```

```python
plt.hist(sine_wave, bins=256)
plt.show()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
n_bins = 4
reshaped_sine_wave = sine_wave.reshape(-1, 1)
encoder = KBinsDiscretizer(
    n_bins=n_bins,
    encode='ordinal',
    strategy='uniform',
    random_state=0,
)
compressed_sine_wave = encoder.fit_transform(reshaped_sine_wave)
```

```python
plt.plot(compressed_sine_wave)
```

```python
plt.hist(compressed_sine_wave, bins=256)
plt.show()
```
